## Ejercicio 5
Realiza un conversor de grados Celsius a Fahrenheit. El conversor tiene que recibir el input del usuario y devolver la conversión.

La conversión de grados sigue la siguiente fórmula: F = (C × 9 / 5) + 32 

```python
def f_a_c():
    cels = int(input('Escribe grados celsius para convertir a farenheit: '))
    faren = (cels * 9/5) + 32
    print(f"Los grados {cels} en celsius son {faren} grados farenheit")
    return faren

f_a_c()
```

```python
def farenheit_to_celsius():
    """
    Funcion para convertir grados celsius a  grados farenheit.
    """
    cels = int(input('Escribe grados celsius para convertir a farenheit: '))
    faren = (cels * 9/5) + 32 # Formula para convertir grados celsius a ferenheit.
    print(f"Los grados {cels} en celsius son {faren} grados farenheit")
    return faren

farenheit_to_celsius()
```
