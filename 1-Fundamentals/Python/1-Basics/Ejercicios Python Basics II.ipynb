## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

````python

def grados_radiantes():
    grados = float(input('Dame grados para pasar a radiantes: '))
    radiantes = grados * (3.1416/180)
    print(radiantes)
    return radiantes

grados_radiantes()
````

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

````python
def area_paralelogramo():
    altura = float(input('Dame una altura en cm para calcular el area: '))
    base = float(input('Dame una base en cm para calcular el area: '))
    area = altura * base
    print(f"The area is {area}cm")
    return area

area_paralelogramo()
````

## Ejercicio 6
Mismo programa que en 5, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
def iguales():
    var_1 = input('Variable 1: ')
    var_2 = input('Variable 2: ')
    var_3 = input('Variable 3: ')
    comparacion1 = var_1 == var_2 and var_1 == var_3
    comparacion2 = var_1 == var_2 or var_1 == var_3 or var_2 == var_3
    print(comparacion1, comparacion2)

iguales()
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10

````python
def suma_10():
    var_1 = float(input('Variable 1: '))
    var_2 = float(input('Variable 2: '))
    suma =  var_1 + var_2
    if suma == 10:
        print('Es 10')
    if suma > 10:
        print("Mayor que 10")
    if suma < 10:
        print('Menor que 10')
    
suma_10()
````

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

1. Pásalo todo a mayúsculas
2. Pásalo todo a minúsculas
3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
4. Crea una lista dividiéndolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

````python
var_string = "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."


print(f"\n 1. Pásalo todo a mayúsculas ->", f"\n {var_string.upper()}" ) # No sé qué formato prefiero
print(f"\n 2. Pásalo todo a minúsculas ->", var_string.lower())
print(f"\n 3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula ->", var_string.title())
print(f"\n 4. Crea una lista dividiéndolo por sus espacios ->", var_string.split(" "))
print(f"\n 5. Sustituye las comas , por puntos y comas ; ->", var_string.replace(",", ";"))
print(f"\n 6. Elimina las a minúsculas ->", var_string.replace("a", ""))
````

## Ejercicio 11

5. Añade otro elemento en la posicion 3 de la lista
6. Crea otra lista con 4 elementos y concaténala a la que ya tenías.

````python
var_3 = [7, 10, -2]

print(f"\n 1. Crea una lista con 3 elementos numéricos ->", f"\n {var_3}")

var_3.append(44)

print(f"\n 2. Añade un cuarto elemento ->", f"\n {var_3}")

var_suma = sum(var_3)

print(f"\n 3. Calcula la suma de todos ->", f"\n {var_suma}")

var_del = var_3.pop(2)

print(f"\n 4. Elimina el segundo elemento de la lista ->", f"\n Elemento borrado: {var_del}", f"\n Lista: {var_3}")

var_3.insert(2, -12)

print(f"\n 5. Añade otro elemento en la posicion 3 de la lista ->", f"\n Añadimos -12: {var_3}")

var_0 = [1, 3, 5, 7]

var_3.extend(var_0)

print(f"\n 6. Crea otra lista con 4 elementos y concaténala a la que ya tenías. ->", f"\n {var_0}", f"\n {var_3}")

var_sorted = sorted(var_3) # Un método para ordenar (Crea una lista nueva, no modifica la original)

print(f"\n 7. Ordena la lista de menor a mayor. ->", f"\n {var_sorted}")

var_3.sort(reverse=True) # Otro método para ordenar (Modifica la lista original)

print(f"\n 8. BONUS: Ordena la lista de mayor a menor. ->", f"\n {var_3}")
````
