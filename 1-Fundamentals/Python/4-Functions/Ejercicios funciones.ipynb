## Ejercicio 1
Escribe una función que convierta números del 1 al 7 en nombres de los dias de la semana. La función constará de un único argumento numérico y una salida de tipo string

````python
def number_day():
    num = input('Dame un numero y te daré un día: ')
    try: 
        num = int(num)
        days = ('monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday')
        dic_days = dict(enumerate(days))
        return (dic_days[num - 1]).capitalize()
    except Exception as e:
        print(e, 'Error')
        
number_day()

````

En el ejercicio 8 de bucles, creábamos una pirámide invertida, cuyo número de pisos venía determinado por un input del usuario. Crear una función que replique el comportamiento de la pirámide, y utiliza un único parámetro de entrada de la función para determinar el número de filas de la pirámide, es decir, elimina la sentencia input.

````python
def piramide():
    n = int(input('Dame un número entero y te haré una pirámide'))
    for x in range(n):
        for y in range(1, n+1, 1):
            print(y, end=' ')
        print()
        n = n-1
    
piramide()
````

## Ejercicio 3
Escibe una función que compare dos números. La función tiene dos argumentos y hay tres salidas posibles: que sean iguales, que el primero se  mayor que el segundo, o que el segundo sea mayor que el primero

````python
def compare_nums(num1, num2):
    if num1 == num2:
        return 'Iguales'
    elif num1 < num2:
        return 'Menor'
    else:
        return 'Mayor'
        

compare_nums(7, 1)
````

## Ejercicio 4
Escribe una función que sea un contador de letras. En el primer argumento tienes que introducir un texto, y el segundo que sea la letra a contar. La función tiene que devolver un entero con el número de veces que aparece esa letra, tanto mayuscula, como minúscula

````python
def count_letter(phrase, letter):
    counter = phrase.count(letter)
    return counter

count_letter('hola, buenos días, como estas', 'a')

````

Escribe una función que tenga un único argumento, un string. La salida de la función tiene que ser un diccionario con el conteo de todas las letras de ese string.

## Ejercicio 6
Escribir una función que añada o elimine elementos en una lista. La función necesita los siguientes argumentos:
* lista: la lista donde se añadirán o eliminarán los elementos
* comando: "add" o "remove"
* elemento: Por defecto es None.

````python
def list_addorremove(list, command, element):
    if command == 'add':
        list.append(element)
    if command == 'remove':
        list.remove(element)
    return list

list_addorremove([1, 2, 3], 'remove', 2)
````

## Ejercicio 7
Crea una función que reciba un número arbitrario de palabras, y devuelva una frase completa, separando las palabras con espacios.

````python
def create_sentence(*args):
    sentence = ' '.join(args)
    
    return sentence

create_sentence('hola', 'compañero', 'buenos', 'días')

````

## Ejercicio 9
Define en una única celda las siguientes funciones:
* Función que calcule el área de un cuadrado
* Función que calcule el area de un triángulo
* Función que calcule el área de un círculo

En otra celda, calcular el area de:
* Dos círculos de radio 10 + un triángulo de base 3 y altura 7
* Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4

````python
from math import pi

def area_cuadrado(lado):
    return lado**2

def area_triangulo(base, altura):
    return (base * altura)/2

def area_circulo(radio):
    return (radio**2)*pi
````

````python
print(area_circulo(10) + area_circulo(10) + area_triangulo(3, 7))
````
